```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
precision = [0.52748, 0.63918, 0.57300, 0.45274]

recall = [0.718160, 0.838336, 0.802144, 0.795472]

modes = [12.0, 19.4, 19.4, 18.6]

ratio = [0.3, 0.5, 0.8, 0.9]
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,3))
axs[0].plot(ratio, precision, label="Precision")
axs[0].plot(ratio, recall, label="Recall")
axs[0].set_xlabel("Ratio")
axs[0].legend()

axs[1].plot(ratio, modes, label="Covered Modes")
axs[1].set_xlabel("Ratio")
axs[1].set_ylim(0, 25)
axs[1].legend()

plt.suptitle("Precision, Recall & Mode Coverage given Phase 1 Ratio")
plt.show()
```
